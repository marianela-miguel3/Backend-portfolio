```python
#imagen en formato BGR
img_new= imread('2133.jpg')
plt.imshow(img_new)
plt.show()
```

```python
#de esta manera se puede invertir tambien de bgr a rgb
img_new_RGB = img_new[:,:,::-1]
plt.imshow(img_new_RGB)
plt.show()
```

```python
img_new.shape
```

```python
x, y = 60, 79  # Coordenadas
pixel_value = img_new[x, y]  # Nota: primero va la fila (y), luego la columna (x)

print(f'Valor del píxel en ({x}, {y}): {pixel_value}')
```

```python
#otra manera de hacer lo mismo
img_new[60,79]
```

```python
img_rgb=cv2.cvtColor(img_new, cv2.COLOR_BGR2RGB)
#otra manera PARA TRASNFORMARLO A RGB
```

```python
plt.imshow(img_rgb)
plt.show()
```

```python
# img_canny= cv2.Canny(img_rgb, 100, 200)
#me detecta los bordes de la imagen
img_canny= cv2.Canny(img_rgb, 50, 100)
plt.imshow(img_canny)
```

```python
#lo que hace es desenfocar
img_blur=cv2.GaussianBlur(img_rgb, (51,1),0)
#el primer valor es para desenfocar en horizontal y el segundo en vertical
#IMPORTANTE: ambos deben ser numeros impares
plt.imshow(img_blur)
plt.show()
#probamos desenfocar en horizontal
```

```python
#ahora probamos en vertical
img_blur=cv2.GaussianBlur(img_rgb, (1,51),0)
plt.imshow(img_blur)
plt.show()
```

```python
#lo suavizo un poco mas para que me muestre una especie de imagen con el formato de retrato que se centra
#solo en la imagen principal y difumina el resto
img_blur=cv2.GaussianBlur(img_rgb, (1,11),0)
plt.imshow(img_blur)
plt.show()
```
